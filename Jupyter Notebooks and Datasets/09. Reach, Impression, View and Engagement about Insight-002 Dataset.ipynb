```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reach, Impression, View and Engagement about Insight-002 Dataset

Importing dataset

```python
insgt2=pd.read_excel('insight_002.xlsx')
```

```python
insgt2=insgt2.drop(0)
```

```python
insgt2 = insgt2[insgt2.columns.drop(list(insgt2.filter(regex='Paid')))]
insgt2 = insgt2.drop(columns=['Posted','Post ID', 'Post Message', 'Type', 'Countries', 'Languages','Audience Targeting'])
```

```python
insgt2
```

```python
insgt2['Total Reach']=insgt2["Lifetime Post Total Reach"] + insgt2["Lifetime Post organic reach"] + insgt2["Lifetime Post reach by people who like your Page"]
insgt2['Total Impressions']=insgt2["Lifetime Post Total Impressions"] + insgt2["Lifetime Post Organic Impressions"]
insgt2['Total Engagements']=insgt2["Lifetime Engaged Users"] + insgt2["Lifetime People who have liked your Page and engaged with your post"]
insgt2['Total Consumptions']=insgt2["Lifetime Matched Audience Targeting Consumptions on Post"] + insgt2["Lifetime Matched Audience Targeting Consumers on Post"]
insgt2['Total Negative Feedbacks']=insgt2["Lifetime Negative Feedback from Users"] + insgt2["Lifetime Negative Feedback"]
insgt2['Total Views']=insgt2["Lifetime Organic views to 95%"] + insgt2["Lifetime Organic views to 95%.1"] + insgt2["Lifetime Organic Video Views"] + insgt2["Lifetime Organic Video Views.1"]
```

Summing the similar datsets

```python
insgt2
```

```python
insgt2=insgt2[insgt2.columns.drop(list(insgt2.filter(regex='Lifetime')))]
insgt2
```

```python
insgt2.loc['sum'] = insgt2.sum()
insgt2
```

```python
Reach = insgt2.at['sum','Total Reach']
Impressions = insgt2.at['sum','Total Impressions']
Engagements = insgt2.at['sum','Total Engagements']
Views = insgt2.at['sum','Total Views']
```

```python
plt.rcParams['font.size'] = 16.0
slices=[Reach, Impressions, Engagements, Views]
labels=['Total Reach','Total Impressions','Total Engagements','Total Views']

plt.pie(slices, labels=labels, autopct='%1.1f%%', shadow=True, radius=2, wedgeprops={'edgecolor': 'white'}, explode=[0,0,0.2,0.2,])

plt.show()
```

### We can see that, the views and engagements are comparatively way lower than the reach and impressions of the page. So, more reach and impression do not always bring more views and engagements.
