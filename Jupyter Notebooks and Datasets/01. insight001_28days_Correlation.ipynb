### Importing packages and Dataset...

```python
import csv
import numpy as np
import pandas as pd
import seaborn as sb
from scipy import stats
import matplotlib.pyplot as plt
```

```python
dpd1=pd.read_csv("D:/MAC 4.2/insight_001.csv")
f=open('D:/MAC 4.2/insight_001.csv')
dc1=list(csv.reader(f,delimiter=','))
```

### Taking the columns which contain data of 28 days and placing them to new dataset named "28days.csv"

```python
headers28=[]
columns=[]
for i in range(1,len(dc1[0])):
    if '28 Days' in dc1[0][i]:
        count=0
        for r in range(2,len(dc1)):
            if dc1[r][i]=='':
                count=count+1
        if count<(len(dc1)-2):
            #print(dc1[0][i])
            headers28.append(dc1[0][i].replace("28 Days ",""))
            columns.append(i)

len(headers28)
```

```python
with open('28daysData.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers28)
    for i in range(2,29):
        rows=[]
        for c in columns:
            rows.append(dc1[i][c])
        writer.writerow(rows)
```

### Now, finding correlation, saving them in different csv for different methods: Pearson, Spearman

```python
d28=pd.read_csv('28daysData.csv')
cT=['pearson','spearman']
for i in cT:
    corrStat = d28.corr(method=i)
    corrStat.to_csv('28days_'+i+'_corr.csv')
```

Now, we check the difference of correlation coefficients between two factors which are found from Pearson & Spearman methods.

Pearson correlation is more suitable for raw data having linear relationship.
Spearman correlation is more suitable for ranked data having either monotonic or linear relationship.

If the difference is higher than 0.19, we won't take that coefficient from any method to avoid ambiguity.
This value is taken such a way that overall discarded data is just below 5%.

```python
f=open('28days_pearson_corr.csv')
pr28=list(csv.reader(f,delimiter=','))
f=open('28days_spearman_corr.csv')
sp28=list(csv.reader(f,delimiter=','))
count=0
denom=0
for i in range(1,len(sp28)):
    for j in range(1,len(sp28)):
        denom=denom+1
        if sp28[i][j]!='':
            if abs(float(sp28[i][j])-float(pr28[i][j]))>0.19:
                sp28[i][j]=''
                count=count+1
            
print(' ',count,'cells from Spearman Correlation dataset are discarded which is ',(count*100)/denom,'% of total data\n')
print(' The discarded data of Spearman have higher difference than those of Pearson, \n so we cannot  confidently say that\n the relationship between two factors is linear or monotonic')
```

### Now, saving the pairs having strong positive & negative correlation in txt file

```python
strngPos=[]
file1=open('28daysDataPosCor.txt', 'w', newline='')
file1.write('Strong Positive Correlation Between\n')
strngNeg=[]
file2=open('28daysDataNegCor.txt', 'w', newline='')
file2.write('Strong/Moderate Negative Correlation Between\n')
for r in range(1,len(sp28)):
    for c in range(r+1,len(sp28)):
        try:
            if float(sp28[r][c])>0.99:
                strngPos.append(sp28[0][r]+' & '+sp28[0][c]+' :  '+sp28[r][c])
                file1.write(str('-'+sp28[0][r]+' & '+sp28[0][c]+' :  '+sp28[r][c]+'\n'))
            elif float(sp28[r][c])<(-0.85):
                strngNeg.append(sp28[0][r]+' & '+sp28[0][c]+' :  '+sp28[r][c])
                file2.write(str('-'+sp28[0][r]+' & '+sp28[0][c]+' :  '+sp28[r][c]+'\n'))
        except: pass
file1.close()
file2.close()
```

```python
print('Strong Positive Correlation Between\n')
for i in strngPos:
    print(i)
```

```python
print('Strong/Moderate Negative Correlation Between\n')
for i in strngNeg:
    print(i)
```
