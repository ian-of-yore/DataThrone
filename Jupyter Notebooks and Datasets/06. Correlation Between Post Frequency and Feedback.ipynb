```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Correlation Between Post Frequency and Feedback

Importing Dataset

```python
feedback = pd.read_csv('feeback.csv')
feedback.head()
```

```python
feedback = feedback.drop('Unnamed: 0', axis = 1)
feedback = feedback.set_index('Date')
feedback.head()
```

Separating negative feedback dataframe

```python
negative = feedback[feedback.columns[0:30]]

negative.head()
```

```python
negative.corr()
```

```python
sns.heatmap(negative.corr())
```

Negative feedback correlation

```python
negative['neg sum']=negative.sum(axis=1)
negative.head()
```

Separating positive feedback dataframe

```python
positive = feedback[feedback.columns[30:66]]

positive.head()
```

```python
positive.corr()
```

```python
sns.heatmap(positive.corr())
```

Negative feedback correlation

```python
positive['pos sum']=positive.sum(axis=1)
positive.head()
```

Importing frequency dataset

```python
freq = pd.read_csv('frequency.csv')
freq.head()
```

```python
freq = freq.drop('Unnamed: 0', axis = 1)
freq = freq.set_index('Date')
freq.head()
```

```python
freq = freq[freq.columns.drop(list(freq.filter(regex='Weekly')))]
freq = freq[freq.columns.drop(list(freq.filter(regex='Daily')))]
freq.head()
```

```python
freq.corr()
```

```python
sns.heatmap(freq.corr())
```

Frequency Correlation

```python
freq['freq sum']=freq.sum(axis=1)
freq.head()
```

```python
freqs = freq[freq.columns.drop(list(freq.filter(regex='28')))]
freqs.head()
```

```python
positives = positive[positive.columns.drop(list(positive.filter(regex='Positive')))]
positives.head()
```

```python
negatives = negative[negative.columns.drop(list(negative.filter(regex='Negative')))]
negatives.head()
```

```python
freq_pos = pd.merge( freqs, positives, left_index=True, right_index=True)
freq_pos = freq_pos.drop(freq_pos.index[22:28])
freq_pos
```

```python
freq_neg = pd.merge( freqs, negatives, left_index=True, right_index=True)
freq_neg = freq_neg.drop(freq_neg.index[22:28])
freq_neg
```

```python
freq_pos_neg = pd.merge( freq_pos, negatives, left_index=True, right_index=True)
freq_pos_neg
```

```python
freq_pos_neg.corr()
```

```python
sns.heatmap(freq_pos_neg.corr())
```

Heatmap of positive, negative feedback and frequency

```python
freq_pos_neg = freq_pos_neg.rename(columns={'freq sum': 'Frequency Distribuition', 'pos sum': 'Positive Feedback', 'neg sum':'Negative Feedback'})
freq_pos_neg
```

```python
sns.jointplot(x='Frequency Distribuition',y='Positive Feedback',data=freq_pos_neg,kind='reg')
```

Positive feedback vs frequency. As the frequency increases the positive feedback gets lower.

```python
sns.jointplot(x='Frequency Distribuition',y='Negative Feedback',data=freq_pos_neg,kind='reg')
```

Negative feedback vs frequency. As the frequency increases the negative feedback gets higher.

```python
pvfreq = freq_pos_neg.pivot_table(values='Frequency Distribuition',index='Positive Feedback',columns='Negative Feedback')
sns.heatmap(pvfreq, cmap='summer')
```

### In the heatmap we can also see that frequency and positive feedback inversely proportional. So, posting more frequently is not more effective.
