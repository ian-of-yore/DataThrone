```python
import pandas as pd 
import seaborn as sns
%matplotlib inline
```

# Data Visualization on Daily Liked, Online and Average Views

Importing Data of Daily Liked and Online

```python
df1 = pd.read_csv('daily_liked_and_online.csv')
df1 = df1.fillna(0)
df1.head()
```

```python
df1 = df1.drop('Unnamed: 0', axis = 1)
df1 = df1.set_index('Date')
df1.head()
```

Taking average of Daily Liked and Online

```python
df1["Average of Liked and Online"] = df1.mean(axis=1)
df1.head()
```

Dropping other columns

```python
df1 = df1[df1.columns.drop(list(df1.filter(regex='Daily')))]
df1.head()
```

Importing Views Data

```python
df2 = pd.read_csv('views.csv') 
df2 = df2.fillna(0)
df2.head()
```

Filtering Daily and Weekly Data

```python
df2 = df2.drop('Unnamed: 0', axis = 1)
df2 = df2.set_index('Date')
df2 = df2[df2.columns.drop(list(df2.filter(regex='Daily')))]
df2 = df2[df2.columns.drop(list(df2.filter(regex='Weekly')))]
df2.head()
```

Taking Average

```python
df2["Average of Views"] = df2.mean(axis=1)
df2.head()
```

```python
df2 = df2[df2.columns.drop(list(df2.filter(regex='28')))]
df2.head()
```

Merging two Data Frames

```python
df3 = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
df3
```

Dropping Last two dates without any Average Data

```python
df3 = df3.drop(df3.index[26:28])
df3
```

Comparing Average Liked and Online vs Average Views

```python
sns.jointplot(x='Average of Liked and Online',y='Average of Views',data=df3,kind='reg')
```

There is inverse relation here between Average Liked and Online vs Average Views. Which is interisting.

```python
sns.pairplot(df3, palette='magma')
```

### Again we can see this inverse trend in views and liked in the pair plots also. So, more people online always does not translate into more views.
