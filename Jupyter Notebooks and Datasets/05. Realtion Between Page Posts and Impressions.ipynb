```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# Data Visualization on Page Posts and Impressions

Importing datasets

```python
df1 = pd.read_csv('impressions.csv')
df1 = df1.fillna(0)

df1.head()
```

```python
df1 = df1.drop('Unnamed: 0', axis = 1)
df1 = df1.set_index('Date')
df1 = df1[df1.columns.drop(list(df1.filter(regex='Daily')))]
df1 = df1[df1.columns.drop(list(df1.filter(regex='Weekly')))]
df1 = df1[df1.columns.drop(list(df1.filter(regex='Paid')))]
df1.head()
```

Keeping only 28 days of data

```python
df2 = pd.read_csv('page_post.csv') 
df2 = df2.fillna(0)
df2.head()
```

```python
df2 = df2.drop('Unnamed: 0', axis = 1)
df2 = df2.set_index('Date')
df2 = df2[df2.columns.drop(list(df2.filter(regex='Daily')))]
df2 = df2[df2.columns.drop(list(df2.filter(regex='Weekly')))]
df2 = df2[df2.columns.drop(list(df2.filter(regex='Paid')))]
df2.head()
```

Marging 2 datasets

```python
df3 = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
df3
```

```python
sns.jointplot(x='28 Days Total Impressions',y='28 Days Reach Of Page Posts',data=df3,kind='reg')
```

Plot of 28 Days Total Impressions vs 28 Days Reach of Page Posts

```python
sns.jointplot(x='28 Days Organic impressions',y='28 Days Organic Reach of Page posts',data=df3,kind='reg')
```

Plot of 28 Days Organnic Impressions vs 28 Days Organic Reach of Page Posts

```python
sns.jointplot(x='28 Days Viral impressions',y='28 Days Viral Reach Of Page Posts',data=df3,kind='reg')
```

Plot of 28 Days Viral Impressions vs 28 Days Viral Reach of Page Posts

```python
sns.pairplot(df3,hue='28 Days Viral impressions',palette='magma')
```

### So, it is clear that relation of viral posts and impressions are proportionally. But organic posts have steady and higher impressions on the pair plot. Viral post translates into more growth. But organic post is more effective in the long run. So, always relying on viral post is not recommended.
