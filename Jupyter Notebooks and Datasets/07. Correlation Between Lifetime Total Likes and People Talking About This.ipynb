Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

Importing Datasets

For the sake of easy-handling, we have made a few chunks of the original given dataset. The newly created chunk of datasets are attached with the zip file.

```python
people_talking = pd.read_csv('people_talking_about_this.csv')
likes = pd.read_csv('likes.csv')
```

```python
people_talking = people_talking.drop(columns=['Unnamed: 0'])
```

```python
df = people_talking.fillna(0)

df.head()
```

Collecting Data related to Age and Gender

```python
new_df = df[df.columns[0:40]]
new_df.head()
```

Considering Only 28 days data since daily and weekly data is not sufficient. 

```python
new_df = new_df[new_df.columns.drop(list(new_df.filter(regex='Daily')))]
new_df.head()
```

```python
new_df = new_df[new_df.columns.drop(list(new_df.filter(regex='Weekly')))]
new_df.head()
```

There are still lots of noises. After brief experimentation, We see that only 4 group of people have significant impact over the dataset. To seperate those four groups we use a threshold value of 200 for the sum of each individual column

```python
new_df.drop([col for col, val in new_df.sum().iteritems() if val < 200], axis=1, inplace=True)
```

```python
new_df
```

Since last six days doesn't contain any values, we are removing those rows.

```python
demo = new_df[:-6]
```

```python
demo_df = demo
demo_df
```

```python
demography_df = demo_df.copy()
```

```python
demography_df.rename(columns = {'28 Days Demographics: People Talking About This - F.18-24':'Talk About This F.18-24 ', 
                        '28 Days Demographics: People Talking About This - F.25-34': 'Talk About This F.25-34',
                       '28 Days Demographics: People Talking About This - M.18-24':'Talk About This M.18-24',
                       '28 Days Demographics: People Talking About This - M.25-34':'Talk About This  M.25-34'}, inplace = True)
```

```python
demography_df
```

Collecting age_wise dataset

```python
age_like = pd.read_csv('age_wise_like.csv')
```

```python
age_like.head()
```

```python
age_like.sum(axis = 0, skipna = True)
```

Only a few of these columns has significant impact over the whole dataset.
A Pie Chart would make this clearer.

```python
total = [1129,  6624, 33254.0, 11108.0, 108536.0, 61789.0]
labels = ['M&F. 13-17', 'M&F. 34+', 'F.18-24', 'F.25-34', 'M.18-24', 'M.25-34']
plt.axis("equal")
plt.pie(total,labels=labels, shadow=True, autopct='%1.1f%%',radius=1.5, explode=[0.2,0.1,0,0.1,0,0])
plt.show()
plt.savefig('LifeTime Like by Age Group.jpg',dpi=200,bbox_inches='tight')
```

Males and Females from age 18 to 34 corresponds to almost 96.5 percent of lifetime total likes. 
So, from here on, taking only this 4 into consideration

```python
age_like.drop([col for col, val in age_like.sum().iteritems() if val < 7000], axis=1, inplace=True)
```

```python
age_like.head()
```

Removing another unrelated feature

```python
age_df = age_like.drop(['Lifetime Total Likes'], axis=1)
age_df
```

```python
age_df.rename(columns = {'Lifetime Likes by Gender and Age - F.18-24':'Total Like - F.18-24', 
                        'Lifetime Likes by Gender and Age - F.25-34': 'Total Like F.25-34',
                       'Lifetime Likes by Gender and Age - M.18-24':'Total Like M.18-24',
                       'Lifetime Likes by Gender and Age - M.25-34':'Total Like  M.25-34'}, inplace = True)
```

```python
age_df = age_df[:-6]
```

```python
age_df
```

To merge these to datasets we need same index/column. Adding Date as index to this total like data

```python
df1 = age_df
df1['Index'] = people_talking['Date']
df1.set_index('Index')
```

```python
df2 = demography_df
df2['Index'] = people_talking['Date']
df2.set_index('Index')
```

```python
merged_df = pd.merge(df1, df2)
merged_df.set_index('Index')
```

# Correlation Between these Features

```python
corrMatrix = merged_df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

Clearly there's strong correlation between Age groups and their corresponding total lifetime likes. To get a clear view, we will plot scatter plot

```python
sns.jointplot(x='Total Like F.25-34',y='Talk About This F.25-34',data=merged_df,kind='reg')
```

```python
sns.jointplot(x='Total Like M.18-24',y='Talk About This M.18-24',data=merged_df,kind='reg')
```

There is some interesting findings here.
As Females total number of likes increases, thier frequency of talking gets lower.
For the Male counterparts, they talk more about this as their total number of likes increases
