Importing Libraries

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
```

Importing Insight_002 Dataset

```python
df = pd.read_excel('insight_002.xlsx')
df.head()
```

Removing Unnecessary Rows

```python
df = df.drop(df.index[0])
```

```python
df = df[df.columns.drop(list(df.filter(regex='Paid')))]
df.head()
```

Dropping Columns that show no significant impact

```python
df = df.drop(['Post ID', 'Post Message', 'Countries',  'Languages', 'Posted', 'Audience Targeting'], axis=1)
```

```python
df.head()
```

```python
df1 = df[df.columns.drop(list(df.filter(regex='Organic views')))]
df1.head()
```

```python
df2 = df1[df1.columns.drop(list(df1.filter(regex='Matched Audience')))]
df2.head()
```

```python
df3 = df2[df2.columns.drop(list(df2.filter(regex='Feedback')))]
df3.head()
```

```python
df4 = df3[df3.columns.drop(list(df3.filter(regex='Video View')))]
df4.head()
```

Renaming Coulumns for Better Understandings

```python
df4.rename(columns = {'Lifetime People who have liked your Page and engaged with your post':'Liketime Liked & Engaged', 
                           'Lifetime Average time video viewed':'Lifetime Average Video Length', 
                              'Lifetime Post reach by people who like your Page':'Lifetime Post Reach by Liked People',
                          'Lifetime Post Impressions by people who have liked your Page': 'Lifetime Post Impression by Liked People',
                          }, inplace = True)
```

```python
df4
```

```python
df = df4
```

```python
df = df.drop(['Lifetime Post organic reach', 'Lifetime Post Organic Impressions'], axis = 1)
```

```python
df
```

```python
multimedia = df.drop(['Lifetime Engaged Users', 'Liketime Liked & Engaged', 'Lifetime Average Video Length', 'Lifetime Video length'], axis = 1)
```

```python
multimedia
```

To have a better understanding of each multimedia types impact on the dataset, We will group them by 'Photos', 'Links', 'Video', 'Status', 'ShareVideo' categories

```python
category = df.groupby("Type")
```

```python
photo = category.get_group('Photo')
photo
```

Removing Insignificant Columns

```python
photo = photo.drop(['Lifetime Post Impression by Liked People', 'Liketime Liked & Engaged', 'Lifetime Average Video Length', 'Lifetime Video length', 'Lifetime Post Reach by Liked People'], axis = 1) 
```

```python
photo
```

Renaming Columns for Better Understandings

```python
photo.rename(columns = {'Lifetime Post Total Reach':'Total Reach by Photo',
                      'Lifetime Post Total Impressions': 'Total Impression by Photo',
                      'Lifetime Engaged Users': 'Total Engaged Users by Photo' }, inplace = True)
```

```python
photo
```

```python
photo = pd.DataFrame(photo.sum(), columns=['sum'])
photo
```

A pie Chart to show the impact of each medium on total reach

```python
reach = [332059,15208,13446,126972,29817]
medium = ["Photo","SharedVideo","Link","Video ","Status"]

plt.axis("equal")
plt.pie(reach,labels=medium, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0.1,0.15,0.4,0.1,0.1])
plt.savefig("Total Reach by Medium.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

A pie Chart to show the impact of each medium on total Impression

```python
impression = [440995,18720,17050,205324,40089]
medium = ["Photo","SharedVideo","Link","Video ","Status"]

plt.axis("equal")
plt.pie(impression,labels=medium, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0.1,0.15,0.4,0.1,0.1], counterclock=True, startangle=45)
plt.savefig("Total Impression by Medium.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

A pie Chart to show the impact of each medium on total Engagement

```python
Engagement = [27067,1066,1120,13853,2523]
medium = ["Photo","SharedVideo","Link","Video ","Status"]

plt.axis("equal")
plt.pie(Engagement,labels=medium, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0.1,0.4,0.4,0.1,0.1])
plt.savefig("Total Engagement by Medium.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

We can observe a very interesting trait. Photo dominates in every sector. But the more interesting trait is how much of percentage it dominates. It's always close to 60%, Video is the second most popular feature, having an average 25 to 30 percent dominance. Status comes in third with about 5.5%.

Sharedvideo and Link is actually a joint winner in this case, with almost everytime both having a percentage about 2.5.
