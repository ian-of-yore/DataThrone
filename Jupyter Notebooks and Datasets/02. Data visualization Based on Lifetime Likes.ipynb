Importing DataFrames.

For the sake of efficiency we have made a few chunks of the original given dataset. The newly created chunk of datasets are attached with the zip file.

# Data Visualization of Lifetime Likes Based on Country

```python
# Collecting Data Related to Countries
country = likes[likes.columns[19:67]]
country = country.fillna(0)
```

```python
country.reset_index(drop=True, inplace=True) # Dropping Index(date)
```

```python
country_df = pd.DataFrame(country.sum(), columns=['sum'])
```

```python
# This will help to collect columns names
country_df = country_df.transpose() 
country_df
```

```python
country_list = country_df.columns.tolist()
```

```python
# Removing Unnecessary Strings

cuntry = []
for i in range(len(country_list)):
           newstr =  country_list[i].replace('Lifetime Likes by Country - ', '')
           cuntry.append(newstr)
```

```python
# Collecting Numerical Country Data corresponding to their countries

country_likes = country_df['sum'].values
country_likes
```

We have found that Bangladesh has tremendous affect on Lifetime Likes over all other nations in the world.

Total Likes by People in Bangladesh: 7227.0

Total Likes by People from the Rest of the World: Bangladesh = 215165.0

```python
BD_vs_Other = [215165.0,7227.0]
labels = ["Bangladesh","All Other Countries"]

plt.axis("equal")
plt.pie(BD_vs_Other,labels=labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.3,])
plt.show()
```

For those 3.2%, we want to see which countries have the most effects.

Considering Countries with more than 100 likes(excluding Bangladesh).

```python
## Excluding countries with less than 100 likes and to exclude Bangladesh, We have set the max threshold value to 1000.0

country_df = country_df[(country_df['sum'] > 100.0) & (country_df['sum'] < 1000.0)]
country_df
```

```python
country_list = country_df.transpose()
```

```python
country_list = country_list.columns.tolist()
```

```python
# Removing Unnecessary strings to make the visualization neat and clean.

selected_countries = []
for i in range(len(country_list)):
           newstr =  country_list[i].replace('Lifetime Likes by Country - ', '')
           selected_countries.append(newstr)
```

```python
selected_country_likes = country_df['sum'].values
```

A Bar chart showing the top countries(excluding BD) that contribute to Total Liketime Likes.

```python
xpos = np.arange(len(selected_countries))

plt.bar(xpos,selected_country_likes, label="Likes")
plt.xticks(xpos,selected_countries)
plt.ylabel("No of Lifetime Likes")
plt.title('Countries with Lifetime Likes more than 100 excluding BD')
plt.legend()
```

# Data Visualization of Lifetime Likes Based on City

```python
## Collecting City Data

city = likes[likes.columns[67:117]]
city = city.fillna(0)
```

```python
city_df = pd.DataFrame(city.sum(), columns=['sum'])
```

```python
city_df
```

Removing the cities that has less than 500 likes and to exclude Dhaka we have set the max threshold as 3000

```python
## Considering all the cities won't generate any usefull data. So Setting A min threshold value of 500 likes
## and a Max Threshold value of 3000 likes.

df = city_df[(city_df['sum']<3000) & (city_df['sum']>500)]
df
```

```python
city_df = df.transpose()
city_df
```

```python
city_list = city_df.columns.tolist()
city_list
```

```python
## Removing Unnecessary Strings to make the Visualizaiton more meaningful

selected_cities = []
for i in range(len(city_list)):
           newstr =  city_list[i].replace('Lifetime Likes by City - ', '')
           selected_cities.append(newstr)
```

```python
selected_cities
```

```python
city_hunter = []

for i in range(len(selected_cities)):
    x = selected_cities[i].split(", ")
    city_hunter.append(x[0])
```

```python
city_hunter
```

```python
selected_city_likes = df['sum'].values
```

```python
selected_city_likes
```

```python
## A Bar chart showing the top cities(excluding Dhaka) that contributed to the total lifetime likes.

xpos = np.arange(len(city_hunter))

plt.bar(xpos,selected_city_likes, label="Likes")
plt.xticks(xpos,city_hunter)
plt.ylabel("No of Lifetime Likes")
plt.title('Cities with lifetime likes more than 500 excluding Dhaka')
plt.legend()
```
