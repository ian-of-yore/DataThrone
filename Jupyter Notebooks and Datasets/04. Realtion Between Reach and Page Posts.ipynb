```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# Data Visualization on Daily, Weekly and 28 Days Reach of Page Posts

Importing Datasets

```python
plt.style.use("seaborn")
plt.rcParams['font.size'] = 14.0
reach = pd.read_csv('Reach.csv')
reach.head()
```

```python
reach = reach.drop('Unnamed: 0', axis = 1)
reach = reach.set_index('Date')
reach = reach[reach.columns.drop(list(reach.filter(regex='Daily')))]
reach.head()
```

```python
reach = reach[reach.columns.drop(list(reach.filter(regex='Weekly')))]
reach.head()
```

```python
reach = reach.fillna(0) #filling NaN with zero
reach.head()
```

```python
reach.loc['sum'] = reach.sum()
```

```python
reach
```

```python
Total_Reach_28 = reach.at['sum','28 Days Total Reach']
Organic_Reach_28 = reach.at['sum','28 Days Organic Reach']
Viral_Reach_28 = reach.at['sum','28 Days Viral Reach']
```

```python
slices=[Total_Reach_28, Organic_Reach_28, Viral_Reach_28]
labels=['28 days Total Reach', '28 Days Organic Reach', '28 Days Viral Reach']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]

plt.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True,  wedgeprops={'edgecolor': 'black'})
plt.title("28 Days of Reach")
plt.show()
```

Pie chart of 28 days of total reach. A hefty chunk of reach consits of organic reach.

```python
post = pd.read_csv('page_post.csv')
post.head()
```

```python
post = post.drop('Unnamed: 0', axis = 1)
post = post.set_index('Date')

post.head()
```

```python
post = post[post.columns.drop(list(post.filter(regex='Daily')))]
post = post[post.columns.drop(list(post.filter(regex='Weekly')))]
post.head()
```

```python
post = post.fillna(0)
post.head()
```

```python
post.loc['sum'] = post.sum()
post
```

```python
Total_Reach_28 = post.at['sum','28 Days Reach Of Page Posts']
Organic_Reach_28 = post.at['sum','28 Days Organic Reach of Page posts']
Viral_Reach_28= post.at['sum','28 Days Viral Reach Of Page Posts']
```

```python
slices=[Total_Reach_28, Organic_Reach_28, Viral_Reach_28]
labels=['28 Days Total Reach of Page Posts', '28 Days Organic Reach of Page Posts', '28 Days Viral Reach of Page Posts']
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title("28 Days Reach of Page Posts")
plt.show()
```

Pie chart of page posts. Here also a hefty chunk of reach consits of organic reach.

```python
post = post[post.columns.drop(list(post.filter(regex='Paid')))]
reach = reach[reach.columns.drop(list(reach.filter(regex='Paid')))]
```

```python
reach_post = pd.merge(reach, post, left_index=True, right_index=True, how='outer')
reach_post
```

```python
sns.jointplot(x='28 Days Total Reach',y='28 Days Reach Of Page Posts',data=reach_post,kind='reg')
```
